```python
import yfinance as yf
import pandas as pd
from datetime import datetime
```

```python
stock = "MBG.DE"     # Mercedes-Benz
market = "^GDAXI"    # DAX Index
start_date = "2017-01-01"
end_date = "2025-01-01"
```

```python
# Data downlode

df_stock = yf.download(stock, start=start_date, end=end_date)
df_market = yf.download(market, start=start_date, end=end_date)
```

```python
df_stock.to_csv("/Users/farhad/Desktop/mercedes-capm-timevarying/data/raw/mbg_raw.csv")
df_market.to_csv("/Users/farhad/Desktop/mercedes-capm-timevarying/data/raw/dax_raw.csv")
```
