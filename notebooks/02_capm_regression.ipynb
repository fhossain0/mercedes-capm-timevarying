```python
import pandas as pd 

Rt= pd.read_csv("/Users/farhad/Desktop/mercedes-capm-timevarying/data/raw/processed/returns_merged.csv",
                 index_col=0,          # First column as an index 
                 parse_dates=True      # convert index to datetime
                )
Rt.head() # Rt is return from index and MBG both side 
```

```python

Y = Rt["MBG"]   # dependent variable (stock return)
X = Rt["DAX"]   # independent variable (market return)
```

# Adding a constent 


```python
import statsmodels.api as sm

X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
```

```python
print(model.summary())

```

Return variance of MBG is explained by the DAX index is 63%, Which is high for daily return.
MGB.DE is strongly connected to german market 



Abnormal return is .02 % perday, but this is not significant as the p value is above the rejection line, the performance is fully explained by the market movements(fail to reject the null which was H = 0)

```python
results = pd.DataFrame({
    "alpha": [model.params["const"]],
    "beta": [model.params["DAX"]],
    "r_squared": [model.rsquared]
})

results.to_csv("/Users/farhad/Desktop/mercedes-capm-timevarying/data/raw/processed/capm_results.csv", index=False)
```
