```python
import pandas as pd

# Load raw data
df_mbg = pd.read_csv("/Users/farhad/Desktop/mercedes-capm-timevarying/data/raw/mbg_raw.csv", index_col= 0, parse_dates=True) # parse_dates to convert the index as an real time and date which column looks like of it.
df_dax = pd.read_csv("/Users/farhad/Desktop/mercedes-capm-timevarying/data/raw/dax_raw.csv", index_col= 0, parse_dates=True) # index_col= 0 consider date and time as an object.

df_mbg.head(), df_dax.head()
```

```python
def get_price(df):
    if "Adj Close" in df.columns:
        return df["Adj Close"]
    else:
        return df["Close"]
```

```python
price_mbg = get_price(df_mbg)
price_dax = get_price(df_dax)


```

```python

df_mbg = df_mbg.iloc[1:] # dropped the first row from the stock 
df_mbg = df_mbg.apply(pd.to_numeric, errors='coerce')
print(df_mbg.head())

```

```python
df_dax = df_dax.iloc[2:] # dropped the first row from the dax index 
df_dax = df_dax.apply(pd.to_numeric, errors='coerce') # errors = means apply nan if someting not converted in numeric
print(df_dax.head())
```

```python
price_mbg= df_mbg["Close"]
price_dax= df_dax["Close"]
```

```python
import numpy as np

stockReturn = np.log(price_mbg / price_mbg.shift(1)) 

indexReturn = np.log(price_dax / price_dax.shift(1))

# Log return for continious compounding growth
```

```python
stockReturn
```

```python
# Merging the dataset 
df_returns = pd.DataFrame({
    "MBG": stockReturn,
    "DAX": indexReturn
}).dropna()
```
